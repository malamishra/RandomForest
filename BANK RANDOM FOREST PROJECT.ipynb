```python
selected_col=['default','housing','loan','deposit']
selected_col
```

```python
for col in selected_col:
    data[col]=data[col].map({"yes":1,"no":0})
```

```python
data.head()
```

```python
dummy_data=pd.get_dummies(["job","marital","eduction","contact","month","poutcome"],drop_first=True,dtype=int)
dummy_data
```

```python
data=pd.get_dummies(data,columns=["job","marital","education","contact","month","poutcome"],drop_first=True,dtype="int64")
data
```

```python
x=data.drop(["deposit"],axis=1)
y=data["deposit"]
x.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123,shuffle=True)
print(f"x training {x_train.shape}")
print(f"y training {y_train.shape}")
print(f"x_testing {x_test.shape}")
print(f"y testing {y_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
bank_rf=RandomForestClassifier()
bank_rf
```

```python
bank_rf.fit(x_train,y_train)
y_predict=bank_rf.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
select_models=[LogisticRegression(),
               DecisionTreeClassifier(max_depth= 30,min_samples_leaf= 1,min_samples_split= 5),
               RandomForestClassifier(max_depth= 30,min_samples_leaf= 1,min_samples_split= 5,n_estimators= 200)]
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc=[]
for model in select_models:
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(f"accuracy {accuracy_score(y_pred,y_test)} of model is {model}")
    acc.append(accuracy_score(y_pred,y_test))
    print(acc)
```

```python
result=[acc]
result
```

```python
pd.DataFrame(data=result,columns=["LG","DTC","RF"], index=["accuracy"])
```

```python
rf=RandomForestClassifier()
rf
```

```python
param_grid={
    'n_estimators':[100,200,300,350], #number of tree in the forest
    'max_depth':[None,10,20,30], # Maximum depth of the tree
    'min_samples_split':[2,5,10], # minimum number of sample required to split on internal node
    'min_samples_leaf':[1,2,4] # minmum number of samples required to be at a leaf node 
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search=GridSearchCV(estimator=rf,param_grid=param_grid,n_jobs= -1)
# n_jobes=-1 for faster computation / parallization?

#fit the model with differen hyper parameters combinations
grid_search.fit(x_train,y_train)

#Get the best hyperparameters
best_params= grid_search.best_params_
best_model= grid_search.best_estimator_
```
